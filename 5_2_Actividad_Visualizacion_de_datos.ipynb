# Insurance dataset

El dataset contiene información demográfica sobre los asegurado en una compañía de seguros:

- age: edad del asegurado principal
- sex: género del asegurado. female o male
- bmi: índice de masa corporal
- children: numero de niños que estan cubiertos con la poliza. 
- smoke: si fuma el beneficiario. yes/no
- region: dónde vive el beneficiario. Estos datos son de Estados Unidos. Regiones disponibles: northeast, southeast, southwest, northwest
- charges: costo del seguro.

Carga el dataset `data/insurance.csv` y haz un análisis gráfico de las variables.

```
# Carga las librerías
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

insur = pd.read_csv('insurance.csv')

```

```
# Carga los datos y muestra los primeros renglones
insur.head()
```

```
# ¿Cuáles son las edades de los clientes de la aseguradora? 
# Haz una gráfica de cómo se distribuye esta variable.

# Tamaño de la imagen
fig = plt.figure(figsize=(9,6))

# Gráfico scatterplot. 
sns.histplot(data=insur, x='age')


# Ejes y título. Colocamos la etiqueta correcta de acuerdo a la orientación.
plt.title('Distribución de las edades de la aseguradora')
plt.xlabel('Edades')
plt.ylabel('Cantidad de personas en esas edades')

```

```
# En general, quien tiene costos más altos de seguro, ¿los fumadores o los no fumadores?
# Haz un gráfico que justifique tu respuesta.

# Tamaño de la imagen
fig = plt.figure(figsize=(7,5))
# Gráfico boxplot
sns.boxplot(data=insur, x='charges', y = 'smoker')
# Ejes y título
plt.title('Boxplot de la evaluación de costos entre fumadores y no fumadores')

# "La diferencia entre los costos de no fumadores y fumadores substancial, los fumadores como podemos observar tienen que pagar a lo minimo significativamente mas que el promedio de los no fumadores"
```

```
# ¿Cómo varia el cargo por seguro dependiendo del número de hijos?
# El número de hijos es una variable numérica, pero como tiene pocos valores distintos,
# la podemos considerar también como una variable cualitativa.

# Tamaño de la imagen
fig, axs= plt.subplots(1,2, figsize=(12, 5))

# Gráfico boxplot
sns.boxplot(data=insur, x='charges', y = 'children', ax=axs[0])
sns.barplot(data=insur, x = 'charges', y='children', estimator=np.mean, ax=axs[1])

# Ejes y título
plt.title('Histograma de la evaluación de costos entre la cantidad de hijos que tiene')


# El cargo por seguro no varía mucho
```

```
# ¿Cuál es la región con más usuarios asegurados? 


fig = plt.figure(figsize=(6,4))

# Gráfico countplot para hacer barras con el número de apariciones de cada género.
sns.countplot(data=insur, x = 'region')

# Ejes y título. Colocamos la etiqueta correcta de acuerdo a la orientación.
plt.title('Observaciones de cada género')
plt.xlabel('Género')
plt.ylabel('Frecuencia')

#La region con mas asegurados es el south east
```

```
# ¿Cómo se relacionan las variables numéricas entre sí?
# Muéstralo en un gráfico.

# Vamos a graficar la matriz de correlación del dataset
insur_corr = insur.corr()

# Para graficar el mapa de calor usamos heatmap. No necesitamos especificar x ni y
sns.heatmap(data=insur_corr)
```

```
# ¿Qué relación hay entre el BMI y la edad? ¿Afecta el género en esta relación?
# Muéstralo en un gráfico y verifica si coincide con lo que obtuviste en la pregunta anterior.



# Tamaño de la imagen
fig = plt.figure(figsize=(9,6))

# Gráfico scatterplot. 
sns.scatterplot(data=insur, x = 'age', y='bmi', hue='sex')

# Ejes y título. Colocamos la etiqueta correcta de acuerdo a la orientación.
plt.title('Relación entre la edad y el BMI por género')
plt.xlabel('Edad')
plt.ylabel('BMI')
```

```
# ¿Cuál es el cargo promedio por género de un seguro?
fig = plt.figure(figsize=(6,4))

# Gráfico countplot para hacer barras con el número de apariciones de cada género.
# Como queremos el promedio, el estimador es np.mean
# Otras opciones son np.sum, np.max, np.min, np.std

sns.barplot(data=insur, x = 'sex', y='charges', estimator=np.mean)

# Ejes y título. Colocamos la etiqueta correcta de acuerdo a la orientación.
plt.title('Observaciones de cada género')
plt.xlabel('Cargo promedio')
plt.ylabel('Género')
```

```
# En una sola imagen, grafica las relaciones que tienen las variables
# age, bmi, smoker y region con charges.

fig, axs= plt.subplots(1,4, figsize=(25, 5))
sns.scatterplot(data=insur, x = 'charges', y='age', ax=axs[0])
sns.scatterplot(data=insur, x = 'charges', y='bmi', ax=axs[1])
sns.barplot(data=insur, y='smoker', x='charges', ax=axs[2])
sns.barplot(data=insur, y='region', x='charges', ax=axs[3])
#plt.tight_layout()

```
